## 爬蟲程式(單一頁面)

```python
#request
url = 'https://www.imdb.com/search/title/?title_type=feature&release_date=1894-01-01,2021-12-31&user_rating=1.0,10.0&countries=us&adult=include'

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'
}

response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.text, "lxml")
```

```python
#建立list們
title_list = []
certificate_list=[]
runtime_list=[]
genre_list=[]
rating_list=[]
metascore_list=[]
text_list=[]
director_list=[]
stars1_list=[]
stars2_list=[]
stars3_list=[]
stars4_list=[]
votes_list=[]
```

```python
#soup撈資料
for elms in soup.select("div.lister-item-content"):
    
    #=標題=================================================
    data1 = elms.select("h3.lister-item-header a")
    try:
        title_list.append(data1[0].get_text())
    except:
        title_list.append("NaN")
        
    #=分級=================================================
    data2 = elms.select('span[class="certificate"]')
    try:
        certificate_list.append(data2[0].get_text())
    except:
        certificate_list.append("NaN")
        
    #=片長=================================================
    data3 = elms.select('span[class="runtime"]')
    try:
        runtime_list.append(data3[0].get_text())
    except:
        runtime_list.append("NaN")
        
    #=類型=================================================
    data4 = elms.select('span[class="genre"]')
    try:
        strContent = data4[0].get_text()
        strContent = strContent.replace("\n", "")
        genre_list.append(strContent)
    except:
        genre_list.append("NaN")
        
    #=評分=================================================
    data5 = elms.select('div[class="inline-block ratings-imdb-rating"]')
    try:
        rating_list.append(data5[0].get("data-value"))
    except:
        rating_list.append("NaN")
        
    #=Metascore=================================================
    data6 = elms.select('div[class="inline-block ratings-metascore"]>span')
    try:
        metascore_list.append(data6[0].get_text())
    except:
        metascore_list.append("NaN")
        
    #=簡介=================================================
    data7 = elms.select('div[class="lister-item-content"]>p:nth-child(4)')
    try:
        strContent1 = data7[0].get_text()
        strContent1 = strContent1.replace("\n", "")
        text_list.append(strContent1)
    except:
        text_list.append("NaN")
        
    #=導演=================================================
    data8 = elms.select('div[class="lister-item-content"]>p[class=""]>a:nth-child(1)')
    try:
        director_list.append(data8[0].get_text())
    except:
        director_list.append("NaN")
        
    #=演員1=================================================
    data9 = elms.select('div[class="lister-item-content"]>p[class=""]>a:nth-child(3)')
    try:
        stars1_list.append(data9[0].get_text())
    except:
        stars1_list.append("NaN")
        
    #=演員2=================================================
    data10 = elms.select('div[class="lister-item-content"]>p[class=""]>a:nth-child(4)')
    try:
        stars2_list.append(data10[0].get_text())
    except:
        stars2_list.append("NaN")
        
    #=演員3=================================================
    data11 = elms.select('div[class="lister-item-content"]>p[class=""]>a:nth-child(5)')
    try:
        stars3_list.append(data11[0].get_text())
    except:
        stars3_list.append("NaN")
        
    #=演員4=================================================
    data12 = elms.select('div[class="lister-item-content"]>p[class=""]>a:nth-child(6)')
    try:
        stars4_list.append(data12[0].get_text())
    except:
        stars4_list.append("NaN")
        
    #=投票數=================================================
    data13 = elms.select('p[class="sort-num_votes-visible"]>span:nth-child(2)')
    try:
        votes_list.append(data13[0].get_text())
    except:
        votes_list.append("NaN")

    
print(title_list[0])
print(certificate_list[0])
print(runtime_list[0])
print(genre_list[0])
print(rating_list[0])
print(metascore_list[0])
print(text_list[0])
print(director_list[0])
print(stars1_list[0])
print(stars2_list[0])
print(stars3_list[0])
print(stars4_list[0])
print(votes_list[0])
```
